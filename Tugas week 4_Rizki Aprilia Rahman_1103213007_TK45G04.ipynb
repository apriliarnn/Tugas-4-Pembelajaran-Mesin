```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
file_path = 'StudentsPerformance.csv'
data = pd.read_csv(file_path)

# Eksplorasi awal data
print("5 Baris Pertama Dataset:")
print(data.head())
print("\nDeskripsi Dataset:")
print(data.describe())
print("\nInformasi Dataset:")
print(data.info())

# Seting visual style untuk seaborn
sns.set(style="whitegrid")

# 1. Distribusi Nilai Matematika, Membaca, dan Menulis
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(data['math score'], kde=True, color='skyblue', bins=20)
plt.title('Distribusi Nilai Matematika')

plt.subplot(1, 3, 2)
sns.histplot(data['reading score'], kde=True, color='salmon', bins=20)
plt.title('Distribusi Nilai Membaca')

plt.subplot(1, 3, 3)
sns.histplot(data['writing score'], kde=True, color='lightgreen', bins=20)
plt.title('Distribusi Nilai Menulis')

plt.tight_layout()
plt.show()

# 2. Rata-rata Nilai Berdasarkan Jenis Kelamin
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='gender', y='math score', ci=None, color='skyblue', estimator=np.mean, label='Math Score')
sns.barplot(data=data, x='gender', y='reading score', ci=None, color='salmon', estimator=np.mean, label='Reading Score')
sns.barplot(data=data, x='gender', y='writing score', ci=None, color='lightgreen', estimator=np.mean, label='Writing Score')
plt.legend()
plt.title('Rata-rata Nilai Berdasarkan Jenis Kelamin')
plt.show()

# 3. Distribusi Nilai Berdasarkan Persiapan Ujian
plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='test preparation course', y='math score', palette='Blues')
plt.title('Distribusi Nilai Matematika Berdasarkan Persiapan Ujian')
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='test preparation course', y='reading score', palette='Reds')
plt.title('Distribusi Nilai Membaca Berdasarkan Persiapan Ujian')
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='test preparation course', y='writing score', palette='Greens')
plt.title('Distribusi Nilai Menulis Berdasarkan Persiapan Ujian')
plt.show()
```
